## Problem 1: Even Number Skipper

```python
for i in range(1, 21, 2):  # Iterate directly over odd numbers
    print(i)
```

## Problem 2: Password Checker

```python
correct_password = "secret123"

while True:
    password_input = input("Please enter the password: ")

    if password_input == correct_password:
        print("Access Granted! Welcome.")
        break  # Exit the loop if password is correct
    else:
        print("Incorrect password. Please try again.")
```

## Problem 3: Number Filter Display

```python
for i in range(1, 11):
    if i == 4:
        continue  # Skip printing 4
    elif i == 8:
        break     # Stop the loop completely at 8
    print(i)      # Default action for all other numbers, including 6
```
